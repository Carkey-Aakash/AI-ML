```python
data = [2,12, 15, 14, 16, 18, 19, 22, 25, 20, 21, 23, 24, 26, 30, 35,
        10, 11, 13, 17, 28, 27, 29, 31, 400]
sorted_data= sorted(data)

def median(lst):
    n= len(lst)
    mid = n // 2
    if n % 2 == 0:
        return (lst[mid-1] + lst[mid]) / 2
    else:
        return lst[mid]

n= len(sorted_data)
mid= n //2 

if n  % 2 == 0:
    lower_half = sorted_data[:mid]
    upper_half = sorted_data[mid:]
else:
    lower_half = sorted_data[:mid]
    upper_half = sorted_data[mid+1:]

Q3 = median(upper_half)
Q1 = median(lower_half)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = []
for num in data:
    if num < lower_bound or num > upper_bound:
        outliers.append(num)


print("Sorted data= ", sorted_data)
print("Q1= ",Q1)
print("Q3= ",Q3)
print("IQR= ", IQR)
print("Outliers present in list= ", outliers)
```

## TASK 2

```python
lst1 = [12, 18, 25, 30, 22, 15, 17, 28, 35, 40, 14, 19, 21, 24, 26, 31, 33, 20, 27, 29]
lst2 = [5, 8, 12, 15, 10, 17, 9, 14, 18, 20, 6, 12, 13, 17, 16, 19, 21, 22, 23, 24]

# Find mean of each list
mean1 = sum(lst1) / len(lst1)
mean2 = sum(lst2) / len(lst2)

# Calculate numerator and denominator so it will be easy for us 
numerator = 0
sum_sq1 = 0
sum_sq2 = 0

for x, y in zip(lst1, lst2):
    diff_x = x - mean1
    diff_y = y - mean2
    numerator += diff_x * diff_y      # sum of product of deviations
    sum_sq1 += diff_x ** 2            # sum of squared deviations of lst1
    sum_sq2 += diff_y ** 2            # sum of squared deviations of lst2

denominator = (sum_sq1 ** 0.5) * (sum_sq2 ** 0.5)

# Step 3: Correlation
r = numerator / denominator

print("Correlation coefficient r =", r)

# Step 4: Interpretation
if r > 0:
    print("The correlation is positive.")
elif r < 0:
    print("The correlation is negative.")
else:
    print("There is zero correlation.")
```
