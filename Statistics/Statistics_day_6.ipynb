```python
import csv

# 1. Load CSV file 

file_path = r"D:\Machine learning\London_Housing_Dataset\londondata.csv" 

with open(file_path, "r", newline='', encoding="utf-8") as f:
    reader = csv.reader(f)
    rows = list(reader)

#  2. Extract header 

header = rows[0]
data = rows[1:]   # all rows except header

#  3. Count missing values per column 

missing_count = {col: 0 for col in header}

for row in data:
    for col_index, value in enumerate(row):
        # consider value missing if empty or only spaces
        if value.strip() == "":
            col_name = header[col_index]
            missing_count[col_name] += 1

#  4. Print the report 
print("=== Missing Value Report ===")
for col in header:
    print(f"{col}: {missing_count[col]} missing values")
```

```python
import csv

# CSV file path
csv_file = r"C:\Users\diamond\Documents\Internship Projects\week 1\Messy_Employee_dataset.csv"
column_name = "Age"  # numeric column to find mode

# 1. Read the column values
values = []

with open(csv_file, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        val = row.get(column_name, "").strip()
        if val != "":
            try:
                values.append(float(val))  # convert to float
            except:
                pass  # ignore invalid/missing values

# 2. Count frequency using a dictionary
freq = {}
for num in values:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

# 3. Find the highest frequency
max_freq = max(freq.values())

# 4. Collect all numbers with the highest frequency (handle multiple modes)
modes = [num for num, count in freq.items() if count == max_freq]

# 5. Print results
print(f"Total valid values in '{column_name}':", len(values))
print("Frequency dictionary:", freq)
print(f"Mode(s) of '{column_name}':", modes)
```

```python
def get_quartile(data, q):
    # Sort the data
    data = sorted(data)
    n = len(data)

    # Position formula for quartile: p = q * (n + 1) / 4
    pos = q * (n + 1) / 4

    # If pos is an integer, return that value directly
    if pos.is_integer():
        return data[int(pos) - 1]

    # Otherwise interpolate
    lower = int(pos) - 1
    upper = lower + 1
    fraction = pos - int(pos)
    return data[lower] + fraction * (data[upper] - data[lower])
    
# Example usage
sample_data = [10, 20, 30, 40, 50, 60, 70]

Q1 = get_quartile(sample_data, 1)
Q2 = get_quartile(sample_data, 2)
Q3 = get_quartile(sample_data, 3)

print("Q1 =", Q1)
print("Median (Q2) =", Q2)
print("Q3 =", Q3)
```

```python
# Example numeric data (replace with your column values)
data = [10, 12, 14, 18, 19, 20, 25, 30]

def calculate_skewness(data):
    n = len(data)

    # ----- Mean -----
    # total = 0
    # for x in data:
    #     total += x
    # mean = total / n
    mean = sum(data) / n

    # ----- Standard Deviation -----
    # First compute variance
    # variance_sum = 0
    # for x in data:
    #     variance_sum += (x - mean) ** 2
    # variance = variance_sum / n
    # std_dev = variance ** 0.5

    variance = sum((x-mean)**2  for x in data) / n
    std_dev = variance ** 0.5

    # ----- Skewness -----
    skew_sum = 0
    for x in data:
        skew_sum += (x - mean) ** 3
    skewness = (skew_sum / n) / (std_dev ** 3)

    return mean, std_dev, skewness


# ---- Use function ----
mean, std_dev, skewness = calculate_skewness(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Skewness:", skewness)

# ---- Interpretation ----
if skewness > 0:
    print("Distribution is positively skewed (right-skewed).")
elif skewness < 0:
    print("Distribution is negatively skewed (left-skewed).")
else:
    print("Distribution is perfectly symmetrical.")
```

```python
data = [5, 12, 17, 18, 26, 28, 33, 35, 42, 48, 51, 55, 60, 61, 70, 72]

def text_histogram(data, bins=6):
    data = sorted(data)
    minimum = min(data)
    maximum = max(data)

    # Step 1: bin width
    bin_width = (maximum - minimum) / bins

    # Step 2: create bins
    ranges = []
    start = minimum
    for _ in range(bins):
        end = start + bin_width
        ranges.append((start, end))
        start = end

    # Step 3: count values in bins
    counts = [0] * bins

    for value in data:
        for i, (low, high) in enumerate(ranges):
            if low <= value < high:
                counts[i] += 1
                break
        else:
            # put the maximum value in the last bin
            if value == maximum:
                counts[-1] += 1

    # Step 4: print histogram
    for (low, high), count in zip(ranges, counts):
        print(f"{int(low)}–{int(high)}: {'*' * count}")

# Run
text_histogram(data, bins=6)
```

```python
# Example data for 3 numeric columns
x = [10, 12, 15, 18, 20]
y = [8,  9, 11, 13, 15]
z = [2,  3,  5,  7,  9]

# Function: mean
def mean(data):
    return sum(data) / len(data)

# Function: standard deviation
def std(data):
    m = mean(data)
    return (sum((x - m) ** 2 for x in data) / len(data)) ** 0.5

# Function: correlation between two lists
def correlation(a, b):
    mean_a = mean(a)
    mean_b = mean(b)
    sd_a = std(a)
    sd_b = std(b)

    num = sum((a[i] - mean_a) * (b[i] - mean_b) for i in range(len(a)))
    den = len(a) * sd_a * sd_b

    return num / den

# Build correlation matrix
matrix = [
    [correlation(x, x), correlation(x, y), correlation(x, z)],
    [correlation(y, x), correlation(y, y), correlation(y, z)],
    [correlation(z, x), correlation(z, y), correlation(z, z)],
]

# Print the matrix in clean format
cols = ["X", "Y", "Z"]
print("     X       Y       Z")
for row_name, row in zip(cols, matrix):
    print(row_name, " ", "  ".join(f"{value:.3f}" for value in row))
```

```python
# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# Formula: mean = ( Σ xi ) / n

def mean(data):
    return sum(data) / len(data)


# Population Variance:
# Formula: variance(x) = ( Σ (xi - mean(x))² ) / n
def variance(data):
    m = mean(data)
    return sum((v - m) ** 2 for v in data) / len(data)


# Population Covariance:
# Formula: covariance(x,y) = ( Σ (xi - mean(x)) * (yi - mean(y)) ) / n
def covariance(x, y):
    mx = mean(x)
    my = mean(y)
    return sum((x[i] - mx) * (y[i] - my) for i in range(len(x))) / len(x)



# Linear Regression:
# Slope (m) = covariance(x,y) / variance(x)
# Intercept (b) = mean(y) - m * mean(x)

def linear_regression(x, y):
    var_x = variance(x)
    cov_xy = covariance(x, y)
    m = cov_xy / var_x
    b = mean(y) - m * mean(x)
    return m, b

# Predicted Y:
# Formula: y_pred = m * x + b

def predict(x, m, b):
    return [m * xi + b for xi in x]

# Mean Squared Error (Population MSE):
# Formula: MSE = ( Σ (yi - y_pred_i)² ) / n

def mse(y, y_pred):
    return sum((y[i] - y_pred[i]) ** 2 for i in range(len(y))) / len(y)
    
# calling
var_x = variance(x)
cov_xy = covariance(x, y)
m, b = linear_regression(x, y)
y_pred = predict(x, m, b)
error = mse(y, y_pred)

#  PRINTING  RESULTS 
print("Population Variance(x):", var_x)
print("Population Covariance(x, y):", cov_xy)
print("Slope (m):", m)
print("Intercept (b):", b)
print("Predicted y:", y_pred)
print("MSE:", error)
```

```python


def percentile(data, p):

    # lets sort data
    sorted_data = sorted(data)

    # now finding index position 0-based index
    n = len(sorted_data)
    pos = (p/100) * (n-1)

    # if pos is decimal we have 2 index lower and upper
    lower = int(pos)
    upper = lower + 1

    # Exception case: if upper index is out of range
    if upper >= n:
        return sorted_data[lower]

    # formula to calculate percentile value = lower + fractional *(upper - lower)
    fraction = pos - lower
    return sorted_data[lower] + fraction * (sorted_data[upper] - sorted_data[lower])



# Basic statistics


def mean(data):
    return sum(data) / len(data)

def median(data):
    d = sorted(data)
    n = len(d)
    mid = n // 2
    if n % 2 == 0:
        return (d[mid-1] + d[mid]) / 2
    return d[mid]

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)

def std_dev(data):
    return variance(data) ** 0.5



# Task 10 – Summary + Visualization


data = [5, 12, 17, 18, 26, 28, 33, 35, 42, 48, 51, 55, 60, 61, 70, 72]

# L, M, H using percentile boundaries
p33 = percentile(data, 33)
p66 = percentile(data, 66)

L = [x for x in data if x < p33]
M = [x for x in data if p33 <= x <= p66]
H = [x for x in data if x > p66]


# Print full summary

print("===== FULL ANALYTICAL SUMMARY =====")
print("Count:", len(data))
print("Mean:", mean(data))
print("Median:", median(data))
print("Min:", min(data))
print("Max:", max(data))
print("Variance:", variance(data))
print("Std Dev:", std_dev(data))
print("33rd Percentile:", p33)
print("66th Percentile:", p66)

print("\n===== DISTRIBUTION ZONES =====")
print(f"L ({len(L)} values): {'#' * len(L)}")
print(f"M ({len(M)} values): {'#' * len(M)}")
print(f"H ({len(H)} values): {'#' * len(H)}")
```
