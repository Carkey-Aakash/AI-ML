# Task 1

```python
# With only python no libraries

data1 = [12, 45, 7, 23, 23, 9, 16, 30, 11, 18, 18, 18, 50, 2, 27]

# Mean
mean1 = sum(data1) / len(data1)

# Median
s = sorted(data1)
mid = len(s) // 2
median1 = (s[mid] if len(s) % 2 == 1 else (s[mid - 1] + s[mid]) / 2) # s[mid] if length is odd 

# Mode (simple counting)
freq = {}
for x in data1:
    freq[x] = freq.get(x, 0) + 1   # checks if x exists in freq if exists return their current value and add 1 else give 0 and add 1 to it.
mode1 = max(freq, key=freq.get)

# Minimun, Maximum, Range
minimum = min(data1)
maximum = max(data1)
range1 = maximum - minimum


print("Data:", data1)
print("Mean:", mean1)
print("Median:", median1)
print("Mode:", mode1)
print("Min:", minimum)
print("Max:", maximum)
print("Range:", range1)
```

```python
# Manual way without using counters 

data2= [random.randint(1,100) for _ in range(30)]
freq={}
for num in data2:
    if num in freq:
        freq[num]+=1
    else:
        freq[num]= 1
        
sorted_freq = dict(sorted(freq.items()))
print("Data:", data2)
print("Frequency (ascending):", sorted_freq)
```

```python
import random
data3 = [random.randint(1, 100) for _ in range(50)]

# Mean
mean3 = sum(data3) / len(data3)

# Variance (population)
variance3 = sum((x - mean3) ** 2 for x in data3) / len(data3)

# Standard deviation
standard_dev3 = variance3 ** 0.5

print("Data:", data3)
print("The Mean is :", mean3)
print("The Variance is :", variance3)
print("The Standard Deviation is :", standard_dev3)

```

```python


data4 = [random.randint(1, 50) for _ in range(40)]

# Histogram
plt.figure(figsize=(10, 4))
plt.hist(data4, bins=10, edgecolor='black')
plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Cumulative Frequency Line Graph
sorted_data = sorted(data4)
cumulative = []
total = 0

for value in sorted_data:
    total += 1
    cumulative.append(total)

plt.figure(figsize=(10, 4))
plt.plot(sorted_data, cumulative, marker='o')
plt.title("Cumulative Frequency Line Graph")
plt.xlabel("Values")
plt.ylabel("Cumulative Frequency")
plt.grid(True)
plt.show()
```

```python
A = [random.randint(10, 50) for _ in range(20)]
B = [random.randint(10, 50) for _ in range(20)]

# Mean
meanA = sum(A) / len(A)
meanB = sum(B) / len(B)

# Standard deviation
varA = sum((x - meanA) ** 2 for x in A) / len(A)
varB = sum((x - meanB) ** 2 for x in B) / len(B)

stdA = varA ** 0.5
stdB = varB ** 0.5


print("A:", A)
print("B:", B)
print("Mean A =", meanA)
print("Mean B =", meanB)
print("Std Dev A =", stdA)
print("Std Dev B =", stdB)

if stdA > stdB:
    print("Dataset A is more spread out.")
elif stdB > stdA:
    print("Dataset B is more spread out.")
else:
    print("Both datasets have equal spread.")
```
