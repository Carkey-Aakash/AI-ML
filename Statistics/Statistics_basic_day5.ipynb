```python
import random

# Parameters
lambda_val = 5
num_events = 2000

# Function to generate a Poisson random variable using inverse transform method
def poisson_sample(lam):
    L = 2.718281828459045 ** (-lam)  # e^-lambda
    k = 0
    p = 1.0
    while p > L:
        p *= random.random()
        k += 1
    return k - 1  # subtract 1 because loop adds extra 1

# Generate 2000 samples
samples = [poisson_sample(lambda_val) for _ in range(num_events)]

# Find the mode
mode_count = {}
for value in samples:
    if value in mode_count:
        mode_count[value] += 1
    else:
        mode_count[value] = 1

most_common = max(mode_count, key=mode_count.get)

print("Most common outcome (mode) is:", most_common)
print("Lambda (mean) is:", lambda_val)
print(samples[1:15])
```

```python
import random
import math

# Step 1: Generate 2000 values from Normal(mean=60, std=10)
values = []
mean = 60
std = 10

for _ in range(2000):
    # Box-Muller method to generate normal value
    u1 = random.random()
    u2 = random.random()
    z = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)

    x = mean + std * z      # convert z to actual normal value
    values.append(x)

# Step 2: Convert each value to a z-score
z_scores = []
for x in values:
    z_value = (x - mean) / std
    z_scores.append(z_value)

# Step 3: Count how many z-scores fall between -1.96 and +1.96
count = 0
for z in z_scores:
    if -1.96 <= z <= 1.96:
        count += 1

print("Total values:", len(z_scores))
print("Values between -1.96 and +1.96:", count)
print("Below is explanation of 95 % Rule")
```

*In a normal distribution, almost all values fall near the mean.

- 68% of values fall within 1 standard deviation

- 95% of values fall within 2 standard deviations (approx 1.96)

- 99.7% fall within 3 standard deviations

So: About 95% of z-scores are between -1.96 and +1.96*

## TASK 2

```python
import math
import random
import matplotlib.pyplot as plt

mean = 20
std = 4
lam = 20
n = 2000

# ---------------------------
# 1. Generate Normal samples
# ---------------------------
normal_samples = []
for _ in range(n):
    u1 = random.random()
    u2 = random.random()
    z = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    x = mean + std * z
    normal_samples.append(x)

# ---------------------------
# 2. Generate Poisson samples (Inverse Transform Method)
# ---------------------------
poisson_samples = []

for _ in range(n):
    L = math.exp(-lam)
    p = 1
    k = 0
    while p > L:
        p *= random.random()
        k += 1
    poisson_samples.append(k - 1)

# ---------------------------
# 3. Plot both histograms on SAME graph
# ---------------------------
plt.hist(normal_samples, bins=20, alpha=0.5, label="Normal (mean=20, std=4)")
plt.hist(poisson_samples, bins=20, alpha=0.5, label="Poisson (lambda=20)")
plt.legend()
plt.title("Normal vs Poisson (2000 samples)")
plt.show()
```

**Explain why Poisson becomes similar to Normal for large lambda.**

- When  is large, the Poisson distribution becomes similar to a Normal distribution because it can be viewed as the sum of many small,
 independent rare events. According to the Central Limit Theorem, such sums tend to follow a Normal distribution. As  increases, the Poisson
 distribution becomes less skewed, more symmetric, and more spread out, making its shape closely resemble the bell curve. Thus, for large 
 (typically   1520), Poisson() can be well-approximated by Normal(, ).

```python
import random
import math

# factorial
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# combination formula
def combination(n, k):
    return factorial(n) / (factorial(k) * factorial(n-k))

# theoretical probability for binomial
def binomial_theoretical(n, p, k):
    return combination(n, k) * (p**k) * ((1-p)**(n-k))

# simulation function
def simulate_binomial_estimation(n, p, trials):
    count_k8 = 0
    
    for _ in range(trials):
        success = 0
        for _ in range(n):
            if random.random() < p:
                success += 1
        if success == 8:
            count_k8 += 1
    
    return count_k8 / trials  # empirical probability

# Given values
n = 20
p = 0.4
trials = 5000
k = 8

empirical = simulate_binomial_estimation(n, p, trials)
theoretical = binomial_theoretical(n, p, k)

print("Empirical Probability of k=8:", empirical)
print("Theoretical Probability of k=8:", theoretical)
```

```python
import random
import math
import matplotlib.pyplot as plt

# Function to generate normal samples using Box-Muller
def normal_distribution(mean, std, size=200):
    samples = []
    for _ in range(size):
        u1 = random.random()
        u2 = random.random()
        z = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
        x = mean + std * z
        samples.append(x)
    return samples

# Generate two distributions
N1 = normal_distribution(30, 3)
N2 = normal_distribution(30, 10)

# Plot both curves
plt.plot(N1, alpha=0.5, label="N1: mean=30, std=3")
plt.plot(N2, alpha=0.5, label="N2: mean=30, std=10")
plt.legend()
plt.title("Comparison of Two Normal Distributions")
plt.show()
```

**Explanation: How Standard Deviation Affects Shape**

- Standard deviation measures spread of data around the mean.

- Small std (3)  data is closer to mean  narrow, tall peak

- Large std (10)  data is more spread out  wide, shorter peak

- Both still centered at mean=30, but the width changes.

```python
import math
import matplotlib.pyplot as plt

# factorial function
def factorial(k):
    result = 1
    for i in range(1, k+1):
        result *= i
    return result

# Poisson probability function
def poisson_prob(lmbda):
    samples = []
    for k in range(11):  # 0 to 10 inclusive
        p = (lmbda**k * math.exp(-lmbda)) / factorial(k)
        samples.append(p)
    return samples

lmbda = 4
poisson_samples = poisson_prob(lmbda)
k_values = list(range(11))  # k = 0 to 10

# Plot PMF
plt.plot(k_values, poisson_samples, marker='o')
plt.title("Poisson Probability Mass Function (PMF), =4")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.grid(True)
plt.show()

# Identify which k has the highest probability
max_prob = max(poisson_samples)
k_max = k_values[poisson_samples.index(max_prob)]
print(f"Highest probability is at k={k_max}, P={max_prob:.4f}")
```
