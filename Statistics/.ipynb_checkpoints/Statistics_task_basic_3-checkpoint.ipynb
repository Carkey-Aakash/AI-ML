## TASK 1

```python
def percentile(data, p):
    
    # lets sort data
    sorted_data = sorted(data)

    # now finding index position 0-based index
    n = len(sorted_data)
    pos= (p/100) * (n-1)

    # if pos is decimal we have 2 index lower and upper
    lower= int(pos)
    upper = lower + 1 

    # Exception case
    if upper >= n:
        return sorted_data[lower]

    # formula to calculate percentile value = lower + fractional *(upper - lower)
    fraction = pos - lower
    return sorted_data[lower] + fraction * (sorted_data[upper] - sorted_data[lower])

    

import random 
data = [4, 55, 18, 74, 72, 16, 84, 52, 60, 56, 57, 82, 85, 22, 84, 85, 67, 95, 54, 28]
print("Data = ", data)
print("Sorted Data= ",sorted(data))
print("For 25th percentile ",percentile(data,25))
print("For 50th percentile ",percentile(data,50))
print("For 75th percentile ",percentile(data,75))
    
```

## TASK 2

```python
def mid_range(data):
    if not data:
        return None 
    #  Let's find minimum and maximum values
    minimum = min(data)
    maximum = max(data)
    
    # Compute mid-range
    mid = (minimum + maximum) / 2
    return mid

# Example usage
data = [10, 12, 15, 20, 100]  # Notice 100 is an extreme value
print("Mid-range:", mid_range(data))

# another example without outliers
data = [10, 12, 15, 20, 30]  
print("Mid-range:", mid_range(data))
```

```python
#Conclusion as we see the Value with and without outliers so we can say that outliers are sensitive to extreme values
```

```python

def percentile(data, p):
    
    # lets sort data
    sorted_data = sorted(data)

    # now finding index position 0-based index
    n = len(sorted_data)
    pos= (p/100) * (n-1)

    lower= int(pos)
    upper = lower + 1 

    # Exception case
    if upper >= n:
        return sorted_data[lower]

    # formula to calculate percentile value = lower + fractional *(upper - lower)
    fraction = pos - lower
    return sorted_data[lower] + fraction * (sorted_data[upper] - sorted_data[lower])


    
def semi_interquartile_range(data):
   
   #  Compute Semi-Interquartile Range (SIQR) = (Q3 - Q1) / 2
   
    Q1 = percentile(data, 25)  # 25th percentile
    Q3 = percentile(data, 75)  # 75th percentile
    siqr = (Q3 - Q1) / 2
    return siqr

data = [4, 55, 18, 74, 72, 16, 84, 52, 60, 56, 57, 82, 85, 22, 84, 85, 67, 95, 54, 28]
print("Data = ", data)
print("Sorted Data= ",sorted(data))
print("For 25th percentile ",percentile(data,25))
print("For 50th percentile ",percentile(data,50))
print("For 75th percentile ",percentile(data,75))
print("Semi-Inter-Quartile-Range  ",semi_interquartile_range(data))
```

## TASK 4

```python
def  mean_absolute_deviation(data):
    # if no data then avoids error
    if not data:
        return None
    # finding mean
    mean = sum(data) / len(data)

    # diving in parts to find MAD
    numerator= 0 
    for x in data:
        diff= abs(x-mean)
        numerator+= diff
    return  numerator / len (data)

import random
data = [random.randint(1,100) for _ in range(20)]
print("The Data is ", data)
print ("The MAD value is ",  mean_absolute_deviation(data))
    
```

## TASK 5

```python
def percentile(data, p):
    
   # Computing percentile 
    data = sorted(data)
    n = len(data)
    pos = (p / 100) * (n - 1)
    lower = int(pos)
    upper = lower + 1
    if upper >= n:
        return data[lower]
    fraction = pos - lower
    return data[lower] + fraction * (data[upper] - data[lower])

def iqr(data):
    # Compute Interquartile Range i.e using 75-25 percentile
    return percentile(data, 75) - percentile(data, 25)

def standard_deviation(data):
    n = len(data)
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    return variance ** 0.5

def compare_dispersion(data1, data2):
    # Compute SD and IQR for both datasets
    sd1 = standard_deviation(data1)
    sd2 = standard_deviation(data2)
    iqr1 = iqr(data1)
    iqr2 = iqr(data2)

    print(f"Dataset 1: SD={sd1:.2f}, IQR={iqr1:.2f}")  # floating point number and f-string
    print(f"Dataset 2: SD={sd2:.2f}, IQR={iqr2:.2f}")

    # Compare SD
    if sd1 > sd2:
        print("Based on SD: Dataset 1 is more dispersed")
    elif sd2 > sd1:
        print("Based on SD: Dataset 2 is more dispersed")
    else:
        print("Based on SD: Both datasets are equally dispersed")

    # Compare IQR
    if iqr1 > iqr2:
        print("Based on IQR: Dataset 1 is more dispersed")
    elif iqr2 > iqr1:
        print("Based on IQR: Dataset 2 is more dispersed")
    else:
        print("Based on IQR: Both datasets are equally dispersed")

import random
data1 =[ random.randint(1,200) for _ in range(20)]
data2 =[ random.randint(1,200) for _ in range(20)]
compare_dispersion(data1,data2)
```

## TASK 6

```python
def rolling_std(data, window):
    n = len(data)
    if window > n:
        return []  # window too big

    result = []
    for i in range(n - window + 1):  # ensures the window does not go past the end of the list
        segment = data[i: i+window]
        mean = sum(segment) / window
        variance = sum((x - mean) ** 2 for x in segment) / window
        sd = variance ** 0.5
        result.append(sd)
    return result

data = [10, 12, 14, 20, 22]
window = 3

print("Data:", data)
print("Rolling SD (window=3):", rolling_std(data, window))
```

## TASK 7

```python
def detect_high_variability_days(data, threshold):
    high_days = set()  # store high variability days without duplicates
    window = 5

    for i in range(len(data) - window + 1):
        segment = data[i:i+window]
        mean = sum(segment) / window
        sd = (sum((x - mean) ** 2 for x in segment) / window) ** 0.5  # standard deviation
        if sd > threshold:
            high_days.update(range(i, i+window))  # add all days in this window

    return sorted(high_days)


import random

data = [random.randint(50, 150) for _ in range(15)]
threshold = 10

print("Data:", data)
print("High variability days:", detect_high_variability_days(data, threshold))
```

## TASK 8

```python
import random
import matplotlib.pyplot as plt

# Step 1: Create dataset of 50 numbers
data = [random.randint(10, 100) for _ in range(50)]

# Step 2: Compute statistics
n = len(data)
mean = sum(data) / n
std = (sum((x - mean)**2 for x in data) / n) ** 0.5  # standard deviation

# Compute IQR (Interquartile Range)
sorted_data = sorted(data)
Q1 = sorted_data[n//4]
Q3 = sorted_data[3*n//4]
IQR = Q3 - Q1

print("Data:", data)
print("Mean:", mean)
print("Std Deviation:", std)
print("IQR:", IQR)

# Step 3: Plot histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

# Step 4: Draw lines for mean and mean ± std
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')         # draw a vertical line->axvline
plt.axvline(mean + std, color='green', linestyle='dashed', linewidth=2, label='Mean + Std')  #others line style are solid , dotted .........
plt.axvline(mean - std, color='orange', linestyle='dashed', linewidth=2, label='Mean - Std')

plt.title("Histogram of Dataset")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## TASK 9

```python
import random
import matplotlib.pyplot as plt


# Dataset A: low spread (values close together)
dataA = [random.randint(45, 55) for _ in range(20)]
# Dataset B: high spread (values spread out)
dataB = [random.randint(20, 100) for _ in range(20)]

# Step 2: Compute statistics
def compute_stats(data):
    n = len(data)
    mean = sum(data)/n
    std = (sum((x - mean)**2 for x in data)/n)**0.5
    sorted_data = sorted(data)
    Q1 = sorted_data[n//4]
    Q3 = sorted_data[3*n//4]
    IQR = Q3 - Q1
    CV = std / mean
    return mean, std, Q1, Q3, IQR, CV

meanA, stdA, Q1A, Q3A, IQR_A, CV_A = compute_stats(dataA)
meanB, stdB, Q1B, Q3B, IQR_B, CV_B = compute_stats(dataB)

print("Dataset A:", dataA)
print("Mean:", meanA, "Std:", stdA, "IQR:", IQR_A, "CV:", CV_A)
print("Dataset B:", dataB)
print("Mean:", meanB, "Std:", stdB, "IQR:", IQR_B, "CV:", CV_B)

# Step 3: Plot boxplot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)  # 1 row, 2 columns, first plot
plt.boxplot([dataA, dataB], labels=['Dataset A','Dataset B'])
plt.title("Boxplot (IQR difference)")

# Step 4: Plot scatter plot
plt.subplot(1,2,2)  # second plot
plt.scatter(range(len(dataA)), dataA, color='blue', label='Dataset A')
plt.scatter(range(len(dataB)), dataB, color='red', label='Dataset B')
plt.title("Scatter Plot (Spread difference)")
plt.legend()

plt.tight_layout()  # automatically adjusts the spacing between subplots and around labels so that everything fits nicely in the figure.
plt.show()
```

## TASK 10

```python
import random
import matplotlib.pyplot as plt

# Step 1: Define rolling_std function
def rolling_std(data, window):
    result = []
    for i in range(len(data) - window + 1):
        segment = data[i:i+window]
        mean = sum(segment)/window
        sd = (sum((x - mean)**2 for x in segment)/window)**0.5
        result.append(sd)
    return result

# Step 2: Create dataset of 40 numbers
data = [random.randint(50, 100) for _ in range(40)]

# Step 3: Set rolling window size and compute rolling std
window = 5
rolling_sd = rolling_std(data, window)

# Step 4: Plot rolling std
plt.plot(range(len(rolling_sd)), rolling_sd, marker='o', color='blue')
plt.title("Rolling Standard Deviation (window=5)")
plt.xlabel("Segment index")
plt.ylabel("Standard Deviation")
plt.grid(True)
plt.show()

# Optional: print dataset and rolling std
print("Data:", data)
print("Rolling Std Dev:",  rolling_sd)
```
